````python
# ════════════════════════════════════════════════════════
#  CONFIG  —  change values here only
# ════════════════════════════════════════════════════════

# ── Folders ──────────────────────────────────────────────
DOCS_FOLDER        = "/content/drive/MyDrive/Colab Notebooks/Constitution"
JUDGEMENTS_FOLDER  = "/content/drive/MyDrive/Colab Notebooks/Judgements"
INDEX_CACHE_DIR    = "/content/drive/MyDrive/Colab Notebooks/constitution_rag_cache"

# ── Models ───────────────────────────────────────────────
EMBED_MODEL  = "all-MiniLM-L6-v2"
GEMINI_MODEL = "gemini-2.5-flash"
GROQ_MODEL   = "llama-3.3-70b-versatile"

# ── Chunking & retrieval ─────────────────────────────────
CHUNK_SIZE    = 1200
CHUNK_OVERLAP = 250
TOP_K         = 18      # increased — now searching two corpora
OCR_MIN_CHARS = 80      # pages with fewer chars trigger OCR fallback

# ── Court name mapping  (code → full name) ────────────────
COURT_NAMES = {
    "LHC" : "Lahore High Court",
    "SHC" : "Sindh High Court",
    "PHC" : "Peshawar High Court",
    "BHC" : "Balochistan High Court",
    "IHC" : "Islamabad High Court",
    "SC"  : "Supreme Court of Pakistan",
    "AHC" : "Azad Kashmir High Court",
    "FSC" : "Federal Shariat Court",
}

print("✅ Config loaded.")
print(f"   Law docs   → {DOCS_FOLDER}")
print(f"   Judgements → {JUDGEMENTS_FOLDER}")
print(f"   Cache      → {INDEX_CACHE_DIR}")
````

````python
# ════════════════════════════════════════════════════════
#  LLM CLIENTS  —  Gemini primary · Groq fallback · cross-model verify
#  Clients are lazy-loaded on first call so Colab Secrets are always
#  read AFTER the user has set os.environ in the setup cell.
# ════════════════════════════════════════════════════════

import os
import google.generativeai as genai
from groq import Groq
from google.colab import userdata          # ← THIS WAS MISSING
from typing import Tuple

# ── Lazy client holders ───────────────────────────────────
_gemini_model = None
_groq_client  = None


def _get_gemini():
    """Initialise Gemini client on first use."""
    global _gemini_model
    if _gemini_model is None:
        key = userdata.get('GEMINI_API_KEY')
        if not key:
            raise RuntimeError(
                "GEMINI_API_KEY not found in Colab Secrets.\n"
                "Go to 🔑 Secrets (left sidebar) and add GEMINI_API_KEY."
            )
        genai.configure(api_key=key)
        _gemini_model = genai.GenerativeModel(GEMINI_MODEL)
        print(f"✅ Gemini ({GEMINI_MODEL}) initialised.")
    return _gemini_model


def _get_groq():
    """Initialise Groq client on first use."""
    global _groq_client
    if _groq_client is None:
        key = userdata.get('GROQ_API')
        if not key:
            raise RuntimeError(
                "GROQ_API not found in Colab Secrets.\n"
                "Go to 🔑 Secrets (left sidebar) and add GROQ_API."
            )
        _groq_client = Groq(api_key=key)
        print(f"✅ Groq ({GROQ_MODEL}) initialised.")
    return _groq_client


print("✅ LLM clients ready (lazy-loaded on first call).")


# ── Callers ───────────────────────────────────────────────

def call_gemini(prompt: str, temperature: float = 0.2,
                max_tokens: int = 2000) -> Tuple[str, bool]:
    try:
        model = _get_gemini()
        resp  = model.generate_content(
            prompt,
            generation_config=genai.types.GenerationConfig(
                temperature=temperature, max_output_tokens=max_tokens)
        )
        return resp.text, True
    except Exception as e:
        print(f"⚠️  Gemini error ({type(e).__name__}): {e}")
        return "", False


def call_groq(prompt: str, temperature: float = 0.2,
              max_tokens: int = 2000, json_mode: bool = False) -> Tuple[str, bool]:
    try:
        client = _get_groq()
        kwargs = dict(
            model       = GROQ_MODEL,
            messages    = [{"role": "user", "content": prompt}],
            temperature = temperature,
            max_tokens  = max_tokens,
        )
        if json_mode:
            kwargs["response_format"] = {"type": "json_object"}
        resp   = client.chat.completions.create(**kwargs)
        result = resp.choices[0].message.content
        print("✅ Groq responded.")
        return result, True
    except Exception as e:
        print(f"⚠️  Groq error ({type(e).__name__}): {e}")
        return "", False


def call_llm(prompt: str, temperature: float = 0.2,
             max_tokens: int = 2000, json_mode: bool = False) -> str:
    text, ok = call_gemini(prompt, temperature, max_tokens)
    if ok:
        return text
    print("🔄 Falling back to Groq …")
    text, ok = call_groq(prompt, temperature, max_tokens, json_mode)
    if ok:
        return text
    raise RuntimeError("Both Gemini and Groq failed.")


# ── Verifier ──────────────────────────────────────────────

_VERIFY_PROMPT = """You are a senior Pakistani legal expert and QA reviewer.

A legal AI assistant generated this answer for a citizen.
Your tasks:
1. Verify all cited section/article numbers exist in Pakistani law.
2. Verify all judgement citations (court, year, case) are internally consistent.
3. Check for missing important sections, rights, or precedents.
4. Correct any factual errors.
5. Rate: EXCELLENT / GOOD / NEEDS_CORRECTION

If EXCELLENT or GOOD → output unchanged.
If NEEDS_CORRECTION → output corrected version.

Original Question: {question}
Generated Answer:
{answer}

Output the final verified answer ONLY.
Last line must be:
[QA_STATUS: EXCELLENT|GOOD|NEEDS_CORRECTION — verified by {verifier}]
"""

def verify_answer(question: str, draft: str, primary_used_gemini: bool) -> str:
    verifier = "Groq" if primary_used_gemini else "Gemini"
    prompt   = _VERIFY_PROMPT.format(question=question, answer=draft, verifier=verifier)
    print(f"\n🔎 Verifying with {verifier} …")

    if primary_used_gemini:
        verified, ok = call_groq(prompt, temperature=0.1, max_tokens=3000)
        if ok:
            print("✅ Verification complete (Groq).")
            return verified
        return draft

    verified, ok = call_gemini(prompt, temperature=0.1, max_tokens=3000)
    if ok:
        print("✅ Verification complete (Gemini).")
        return verified

    print("⚠️  Gemini unavailable — Groq self-review …")
    sr_prompt = prompt.replace(f"verified by {verifier}", "self-reviewed by Groq")
    verified, ok = call_groq(sr_prompt, temperature=0.1, max_tokens=3000)
    return verified if ok else draft
````

````python
# ════════════════════════════════════════════════════════
#  DOCUMENT PROCESSOR
#  — text extraction with OCR fallback for image PDFs
#  — section header detection (same as before)
#  — judgement metadata extraction (court / year / case title)
# ════════════════════════════════════════════════════════

import re
from pathlib import Path
from typing import List, Optional, Dict

import pdfplumber
import pytesseract
from PIL import Image
from pdf2image import convert_from_path


# ── Court / year parser ──────────────────────────────────────────────────────

def parse_judgement_folder(folder_name: str) -> Dict:
    """
    'LHC2026'  →  {court_code:'LHC', court_name:'Lahore High Court', year:'2026'}
    Handles unknown codes gracefully.
    """
    m = re.match(r'^([A-Z]+)(\d{4})$', folder_name.strip())
    if m:
        code = m.group(1)
        year = m.group(2)
        return {
            "court_code" : code,
            "court_name" : COURT_NAMES.get(code, f"{code} Court"),
            "year"       : year,
        }
    return {"court_code": folder_name, "court_name": folder_name, "year": "Unknown"}


def extract_case_title(first_page_text: str) -> str:
    """
    Try to pull the case title from the top of the first page.
    Looks for 'X vs/v. Y' pattern first, then falls back to first long line.
    """
    lines = [l.strip() for l in first_page_text.split("\n") if l.strip()]
    # Pattern: something v/vs/versus something
    for line in lines[:30]:
        if re.search(r'\bv(?:s\.?|ersus)?\b', line, re.IGNORECASE) and 15 < len(line) < 200:
            return line[:150]
    # Fallback: first line that looks like a title (not a page number / date)
    for line in lines[:10]:
        if len(line) > 20 and not re.match(r'^[\d\s\-/]+$', line):
            return line[:150]
    return "Unknown Case"


# ── OCR fallback ─────────────────────────────────────────────────────────────

def _ocr_page(pdf_path: str, page_number: int) -> str:
    """Convert a single PDF page to image then run Tesseract OCR."""
    try:
        images = convert_from_path(
            pdf_path, first_page=page_number, last_page=page_number, dpi=200
        )
        if images:
            return pytesseract.image_to_string(images[0], lang="eng")
        return ""
    except Exception as e:
        print(f"      ⚠️  OCR failed for page {page_number}: {e}")
        return ""


# ── PDF page extractor ────────────────────────────────────────────────────────

def extract_pages(pdf_path: str,
                  source_meta: Optional[Dict] = None) -> List[Dict]:
    """
    Extract pages from a PDF.
    source_meta — dict with court_code / court_name / year / case_title / source_type.
    Falls back to OCR for pages whose text is below OCR_MIN_CHARS.
    """
    pages     = []
    doc_name  = Path(pdf_path).stem
    meta      = source_meta or {}
    ocr_count = 0

    with pdfplumber.open(pdf_path) as pdf:
        total = len(pdf.pages)
        for i, page in enumerate(pdf.pages, start=1):
            text = (page.extract_text() or "").strip()

            # OCR fallback for image-based pages
            if len(text) < OCR_MIN_CHARS:
                text = _ocr_page(pdf_path, i).strip()
                if len(text) >= OCR_MIN_CHARS:
                    ocr_count += 1

            if len(text) > 30:
                entry = {"page": i, "text": text, "source": doc_name}
                entry.update(meta)   # add court/year/case info if present
                pages.append(entry)

    ocr_note = f" | {ocr_count} OCR pages" if ocr_count else ""
    print(f"   📄 '{Path(pdf_path).name}' → {len(pages)}/{total} pages{ocr_note}")
    return pages


# ── Text pre-processor ────────────────────────────────────────────────────────

def preprocess_text(text: str) -> str:
    """Normalise all section-number variant formats before header detection."""
    text = re.sub(r'(\d{1,4}[A-Z]?)-\s*\n\s*([A-Z])\.\s*',
                  lambda m: f"{m.group(1)}-{m.group(2)}. ", text)
    text = re.sub(r'(\d{1,4}[A-Z]?)-\s*\n\s*([A-Z])\.\s*\n',
                  lambda m: f"{m.group(1)}-{m.group(2)}. ", text)
    text = re.sub(r'\b(\d{1,4}[A-Z]?)/([A-Z])\.',
                  lambda m: f"{m.group(1)}-{m.group(2)}.", text)
    text = re.sub(r'\b(\d{1,4})\s([A-Z])\.\s+([A-Z])',
                  lambda m: f"{m.group(1)}-{m.group(2)}. {m.group(3)}", text)
    text = re.sub(r'(?i)Section\s*(\d{1,4})', r'\1', text)
    return text


# ── Section header detector ───────────────────────────────────────────────────

_SECTION_PATTERN = re.compile(
    r'^(?:(?:Section|Article|Art\.)\s*)?'
    r'(\d{1,4}(?:[.\-/]?[A-Z\d])*)\.\s{1,8}'
    r'([A-Z\u0600-\u06FF][^\n]{2,150})$',
    re.UNICODE,
)
_PART_CHAPTER_PATTERN = re.compile(
    r'^(PART|CHAPTER|SCHEDULE|ANNEX(?:URE)?)\s+[IVXLCDM\d]+',
    re.IGNORECASE,
)


def detect_article_header(line: str) -> Optional[str]:
    line = line.strip()
    if not line:
        return None
    m = _SECTION_PATTERN.match(line)
    if m:
        sec_num = m.group(1)
        title   = m.group(2).strip()
        prefix  = "Article" if re.search(r'[A-Z]', sec_num) and not sec_num.isdigit() else "Section"
        return f"{prefix} {sec_num}: {title}"
    if _PART_CHAPTER_PATTERN.match(line):
        return line.upper().strip()
    return None


# ── Chunker ───────────────────────────────────────────────────────────────────

def chunk_pages(pages: List[Dict]) -> List[Dict]:
    """
    Split pages into overlapping chunks.
    All metadata from the page dict (court, year, case_title, source_type) is
    carried forward into each chunk.
    """
    chunks          = []
    chunk_id        = 0
    current_article = None

    for page_dict in pages:
        page_num  = page_dict["page"]
        # Carry forward all metadata keys except text
        page_meta = {k: v for k, v in page_dict.items() if k not in ("page", "text")}
        raw_text  = preprocess_text(page_dict["text"])
        buffer    = ""

        for line in raw_text.split("\n"):
            header = detect_article_header(line)
            if header:
                if buffer.strip():
                    chunks.append({
                        "chunk_id": chunk_id, "page": page_num,
                        "article": current_article,
                        "text": buffer.strip(), **page_meta
                    })
                    chunk_id += 1
                current_article = header
                buffer = header + "\n"
            else:
                buffer += line + "\n"
                if len(buffer) >= CHUNK_SIZE:
                    chunks.append({
                        "chunk_id": chunk_id, "page": page_num,
                        "article": current_article,
                        "text": buffer.strip(), **page_meta
                    })
                    chunk_id += 1
                    overlap = buffer[-CHUNK_OVERLAP:]
                    buffer  = (f"{current_article}\n" if current_article else "") + overlap

        if buffer.strip():
            chunks.append({
                "chunk_id": chunk_id, "page": page_num,
                "article": current_article,
                "text": buffer.strip(), **page_meta
            })
            chunk_id += 1

    return chunks

print("✅ document_processor ready (text + OCR fallback).")
````

````python
# ════════════════════════════════════════════════════════
#  INDEXER
#  — scans DOCS_FOLDER (flat) + JUDGEMENTS_FOLDER (nested)
#  — smart incremental cache keyed by file MD5
#  — builds one combined FAISS index
# ════════════════════════════════════════════════════════

import hashlib, json, pickle
from pathlib import Path
from typing import Dict, List, Tuple

import faiss
import numpy as np
from sentence_transformers import SentenceTransformer


# ── Cache paths ───────────────────────────────────────────

def _cp_index()  : return f"{INDEX_CACHE_DIR}/index.faiss"
def _cp_chunks() : return f"{INDEX_CACHE_DIR}/chunks.pkl"
def _cp_meta()   : return f"{INDEX_CACHE_DIR}/meta.json"

def _cache_exists():
    return all(Path(p()).exists() for p in [_cp_index, _cp_chunks, _cp_meta])


# ── File fingerprinting ───────────────────────────────────

def _md5(path: str) -> str:
    h = hashlib.md5()
    with open(path, "rb") as f:
        for block in iter(lambda: f.read(65536), b""):
            h.update(block)
    return h.hexdigest()


# ── Folder scanners ───────────────────────────────────────

def scan_law_folder(folder: str) -> Dict[str, Dict]:
    """
    Flat folder (Constitution / PPC).
    Returns {pdf_path: {md5, source_type, ...}}
    """
    p = Path(folder)
    if not p.exists():
        raise FileNotFoundError(f"Law folder not found: {folder}")
    pdfs = sorted(p.glob("*.pdf"))
    print(f"\n📁 Law folder — {len(pdfs)} PDF(s):")
    result = {}
    for pdf in pdfs:
        print(f"   • {pdf.name}")
        result[str(pdf)] = {
            "md5"         : _md5(str(pdf)),
            "source_type" : "law",
        }
    return result


def scan_judgements_folder(folder: str) -> Dict[str, Dict]:
    """
    Nested folder  Judgements/LHC2026/*.pdf
    Returns {pdf_path: {md5, source_type, court_code, court_name, year}}
    """
    p = Path(folder)
    if not p.exists():
        print(f"⚠️  Judgements folder not found: {folder} — skipping.")
        return {}

    result = {}
    sub_folders = sorted([d for d in p.iterdir() if d.is_dir()])
    print(f"\n📁 Judgements folder — {len(sub_folders)} sub-folder(s):")

    for sub in sub_folders:
        court_meta = parse_judgement_folder(sub.name)
        pdfs       = sorted(sub.glob("*.pdf"))
        print(f"   [{sub.name}] {court_meta['court_name']} {court_meta['year']} — {len(pdfs)} PDF(s)")
        for pdf in pdfs:
            result[str(pdf)] = {
                "md5"         : _md5(str(pdf)),
                "source_type" : "judgement",
                **court_meta,
            }
    return result


def scan_all_folders() -> Dict[str, Dict]:
    law        = scan_law_folder(DOCS_FOLDER)
    judgements = scan_judgements_folder(JUDGEMENTS_FOLDER)
    combined   = {**law, **judgements}
    print(f"\n📊 Total files: {len(combined)} ({len(law)} law + {len(judgements)} judgements)")
    return combined


# ── FAISS builder ─────────────────────────────────────────

def _build_faiss(all_chunks: List[Dict], embedder) -> faiss.Index:
    texts = [c["text"] for c in all_chunks]
    print(f"⚙️  Embedding {len(texts)} chunks …")
    emb = embedder.encode(texts, batch_size=64, show_progress_bar=True,
                          convert_to_numpy=True, normalize_embeddings=True)
    idx = faiss.IndexFlatIP(emb.shape[1])
    idx.add(emb.astype(np.float32))
    print(f"✅ FAISS ready — {idx.ntotal} vectors.")
    return idx


# ── Cache helpers ─────────────────────────────────────────

def _save_cache(index, chunks, file_map):
    Path(INDEX_CACHE_DIR).mkdir(parents=True, exist_ok=True)
    faiss.write_index(index, _cp_index())
    with open(_cp_chunks(), "wb") as f: pickle.dump(chunks, f)
    with open(_cp_meta(),   "w")  as f:
        json.dump({"embed_model": EMBED_MODEL,
                   "file_hashes": {p: v["md5"] for p, v in file_map.items()}},
                  f, indent=2)
    print(f"💾 Cache saved — {len(chunks)} chunks, {len(file_map)} files.")

def _load_cache():
    idx = faiss.read_index(_cp_index())
    with open(_cp_chunks(), "rb") as f: chunks = pickle.load(f)
    with open(_cp_meta())   as f: meta = json.load(f)
    return idx, chunks, meta.get("file_hashes", {})


# ── Main builder ──────────────────────────────────────────

def build_or_update_index():
    """
    Build or incrementally update the combined FAISS index.
    Returns (index, embedder, all_chunks).
    """
    print(f"\n🤖 Loading embedder '{EMBED_MODEL}' …")
    embedder    = SentenceTransformer(EMBED_MODEL)
    current_map = scan_all_folders()   # {path: {md5, source_type, ...}}

    # ── Incremental check ─────────────────────────────────
    if _cache_exists():
        print("\n📂 Existing cache found — checking for changes …")
        old_index, old_chunks, cached_hashes = _load_cache()

        new_files     = {p for p in current_map if p not in cached_hashes}
        changed_files = {p for p in current_map
                         if p in cached_hashes and cached_hashes[p] != current_map[p]["md5"]}
        removed_files = {p for p in cached_hashes if p not in current_map}
        stale         = new_files | changed_files | removed_files

        if not stale:
            print("✅ All files up-to-date. Using cached index.")
            return old_index, embedder, old_chunks

        if new_files:     print(f"   ➕ New     : {[Path(p).name for p in new_files]}")
        if changed_files: print(f"   ✏️  Changed : {[Path(p).name for p in changed_files]}")
        if removed_files: print(f"   ❌ Removed : {[Path(p).name for p in removed_files]}")

        surviving = [c for c in old_chunks
                     if not any(c["source"] in Path(p).stem for p in stale)]
        print(f"   ♻️  Kept {len(surviving)} chunks from unchanged files.")
    else:
        print("\n🔨 No cache — building from scratch …")
        surviving = []
        stale     = set(current_map.keys())

    # ── Process stale files ───────────────────────────────
    new_chunks = []
    to_process = {p: v for p, v in current_map.items() if p in stale}
    print(f"\n📥 Processing {len(to_process)} file(s):")

    for pdf_path, meta in to_process.items():
        # Build per-file source_meta (exclude md5)
        source_meta = {k: v for k, v in meta.items() if k != "md5"}

        # For judgements: extract case title from first page before chunking
        if meta["source_type"] == "judgement":
            pages = extract_pages(pdf_path, source_meta=source_meta)
            if pages:
                source_meta["case_title"]  = extract_case_title(pages[0]["text"])
                source_meta["pdf_filename"] = Path(pdf_path).stem
                # Update pages with case_title now known
                for pg in pages:
                    pg["case_title"]   = source_meta["case_title"]
                    pg["pdf_filename"] = source_meta["pdf_filename"]
        else:
            pages = extract_pages(pdf_path, source_meta=source_meta)

        chunks = chunk_pages(pages)
        print(f"      → {len(chunks)} chunks | '{Path(pdf_path).name}'")
        new_chunks.extend(chunks)

    all_chunks = surviving + new_chunks
    for i, c in enumerate(all_chunks):
        c["chunk_id"] = i

    new_index = _build_faiss(all_chunks, embedder)
    _save_cache(new_index, all_chunks, current_map)
    return new_index, embedder, all_chunks

print("✅ Indexer ready.")
````

````python
# ════════════════════════════════════════════════════════
#  RETRIEVAL  (v2)
#  — scenario-aware query expansion (maps everyday situations to legal terms)
#  — FAISS search + keyword injection
#  — relevance cross-check
#  — priority sort: Constitution → PPC → other law → judgements
# ════════════════════════════════════════════════════════

import json, re
from typing import List, Dict
import numpy as np

_index    = None
_embedder = None
_chunks   = None

def set_retrieval_state(index, embedder, chunks):
    global _index, _embedder, _chunks
    _index    = index
    _embedder = embedder
    _chunks   = chunks

_SEC_NUM_RE = re.compile(r'\b(\d{1,4}(?:[.\-/]?[A-Z\d])*)\b', re.UNICODE)


def _source_priority(chunk: Dict) -> int:
    """Lower = retrieved first when building the prompt."""
    if chunk.get("source_type") == "judgement":
        return 3
    name = chunk.get("source", "").lower()
    if "constitution" in name: return 0
    if any(k in name for k in ["penal", "ppc"]): return 1
    return 2


# ── Scenario → Legal Concept mapping prompt ───────────────
# This is the KEY fix: a two-step expansion.
# Step 1: identify if the question is a real-life scenario
# Step 2: map it to Pakistani law concepts + likely section numbers

_EXPAND_PROMPT = """You are a Pakistani legal expert specialising in PPC, Constitution, CrPC, and court judgements.

USER QUESTION: "{question}"

Your job is to rewrite this into a precise legal search query that will find the most relevant law sections and judgements.

STEP 1 — Classify the question:
- "scenario": user describes a real-life situation (e.g., police asking for bribe, arrested without warrant, land dispute)
- "direct": user is asking about a specific law/section/article by number or name

STEP 2 — Map to legal concepts:
Common scenario → law mappings (use these as guidance):
• Police bribery / asking for bribe → PPC Section 161 162 163 165, Prevention of Corruption Act 1947, illegal gratification, public servant
• Arrested without warrant → CrPC Section 54 61, Constitution Article 9 10 10A, unlawful detention, fundamental rights
• Phone / property seized by police → CrPC Section 100 165, Constitution Article 14, search without warrant
• False FIR / case → PPC Section 182 211, malicious prosecution, quashing FIR
• Right to lawyer / advocate → Constitution Article 10A, fair trial, legal representation
• Defamation / social media abuse → PPC Section 499 500 501 502, PECA 2016 Section 20
• Blasphemy → PPC Section 295 295A 295B 295C, religious offences
• Land grabbing / property theft → PPC Section 447 441, Specific Relief Act
• Domestic violence / assault → PPC Section 337 354, Protection of Women Act 2006
• Child abuse → PPC Section 377 328, Zainab Alert Act 2021
• Bail rights → CrPC Section 496 497 498, Constitution Article 9 10
• Cyber crime → PECA 2016, Electronic Crimes Act
• Right to information → Constitution Article 19 19A
• Torture / police brutality → Pakistan Torture Act 2022, Constitution Article 14

STEP 3 — Build search query:
Combine: likely section numbers + legal terms + relevant act name.
Keep query under 25 words.

Reply ONLY in this exact JSON format:
{{
  "question_type": "scenario" or "direct",
  "scenario_summary": "brief 5-word description of the situation",
  "search_query": "the legal search query combining relevant section numbers and legal terms",
  "legal_concepts": ["concept1", "concept2", "concept3"],
  "likely_sections": ["Section 161 PPC", "Section 162 PPC"]
}}"""


def expand_query(user_question: str) -> str:
    """
    Two-step expansion:
    1. Classify question as scenario vs. direct
    2. Map scenario to relevant law sections + legal terms
    Returns the expanded search query string.
    """
    prompt = _EXPAND_PROMPT.format(question=user_question)
    try:
        raw  = call_llm(prompt, temperature=0.1, max_tokens=400, json_mode=True)
        raw  = raw.strip().replace("```json","").replace("```","").strip()
        data = json.loads(raw)

        q_type   = data.get("question_type", "direct")
        expanded = data.get("search_query", user_question)
        concepts = data.get("legal_concepts", [])
        sections = data.get("likely_sections", [])
        summary  = data.get("scenario_summary", "")

        print(f"🔍 Question type : {q_type.upper()}" + (f" — '{summary}'" if summary else ""))
        print(f"🔍 Legal query   : '{expanded}'")
        if concepts: print(f"⚖️  Concepts      : {', '.join(concepts)}")
        if sections: print(f"📋 Likely sections: {', '.join(sections)}")

        return expanded

    except Exception as e:
        print(f"⚠️  Query expansion failed ({e}) — using original.")
        return user_question


def retrieve_chunks(query: str, original_question: str = "") -> List[Dict]:
    """
    Semantic search + keyword injection, sorted by source priority.
    Also injects chunks for any section numbers found in the expanded query
    (not just the original question) — this is critical for scenario questions
    where section numbers only appear after expansion.
    """
    q_vec = _embedder.encode(
        [query], normalize_embeddings=True, convert_to_numpy=True
    ).astype(np.float32)
    scores, ids = _index.search(q_vec, TOP_K)

    results  = []
    seen_ids = set()
    for score, idx in zip(scores[0], ids[0]):
        if idx != -1:
            c = _chunks[idx].copy()
            c["score"] = float(score)
            results.append(c)
            seen_ids.add(idx)

    # ── Keyword injection ─────────────────────────────────
    # Search the expanded query AND original question for section numbers.
    # Scenario questions will have section numbers in the expanded query only.
    combined_text = original_question + " " + query
    all_nums      = set(_SEC_NUM_RE.findall(combined_text))

    expanded_nums = set()
    for sec in all_nums:
        expanded_nums.add(sec)
        for letter in "ABCDE":
            expanded_nums.update([f"{sec}-{letter}", f"{sec}{letter}"])

    injected = 0
    for sec in expanded_nums:
        hits = [c for i, c in enumerate(_chunks)
                if i not in seen_ids and c.get("article") and sec in c["article"]]
        for c in hits[:3]:
            c = c.copy(); c["score"] = 0.55
            results.append(c)
            seen_ids.add(c["chunk_id"])
            print(f"📎 Injected: [{c['source']}] {c['article']} p.{c['page']}")
            injected += 1

    if injected:
        print(f"📎 Total injected: {injected} chunk(s) via section-number matching.")

    results.sort(key=lambda c: (_source_priority(c), -c["score"]))
    return results


def crosscheck_chunks(user_question: str, retrieved: List[Dict]) -> List[Dict]:
    """
    Cross-check retrieved chunks against the user's question.
    For scenario questions, we keep chunks that are relevant to the SITUATION,
    not just those that mention the exact section number.
    """
    if not retrieved:
        return []

    summary = ""
    for i, c in enumerate(retrieved, 1):
        label   = c.get("article") or "General Text"
        src     = c.get("source", "")
        preview = c["text"][:200].replace("\n", " ")
        summary += f"Chunk {i} [{label} | {src} | p.{c['page']}]: {preview}\n\n"

    prompt = f"""A Pakistani citizen asked: "{user_question}"

Below are retrieved legal text chunks. Mark each as RELEVANT if it:
- Directly defines rights or offences related to the question
- Gives procedures the citizen or police must follow
- Is a court ruling on a similar situation
- Defines penalties for related conduct

Mark as NOT_RELEVANT only if the chunk is completely unrelated to the question or situation.
Be INCLUSIVE — for scenario questions, keep any chunk that could help answer the citizen's practical question.

{summary}

Reply ONLY with JSON: {{"relevant_chunks": [1, 3, 5]}}"""

    try:
        raw  = call_llm(prompt, temperature=0, max_tokens=150, json_mode=True)
        raw  = raw.strip().replace("```json","").replace("```","").strip()
        data = json.loads(raw)
        ids      = data.get("relevant_chunks", list(range(1, len(retrieved)+1)))
        filtered = [retrieved[i-1] for i in ids if 1 <= i <= len(retrieved)]
        removed  = len(retrieved) - len(filtered)
        if removed: print(f"🧹 Cross-check removed {removed} irrelevant chunk(s).")
        return filtered if filtered else retrieved
    except Exception as e:
        print(f"⚠️  Cross-check failed ({e}) — keeping all.")
        return retrieved

print("✅ Retrieval module ready.")
````

````python
# ════════════════════════════════════════════════════════
#  ANSWER GENERATOR  (v2)
#  — always cites retrieved chunks even for general questions
#  — clean 3-section output: Answer | Law Refs | Judgement Refs
#  — cross-model verification
# ════════════════════════════════════════════════════════

from typing import List, Dict

# ═══════════════════════════════════════════════════════
#  SYSTEM PROMPT
# ═══════════════════════════════════════════════════════

_SYSTEM_PROMPT = """You are a helpful Pakistani legal assistant for ordinary citizens.
You have been given RETRIEVED PASSAGES from two source types:
  (A) STATUTORY LAW — Constitution of Pakistan, Pakistan Penal Code (PPC), CrPC
  (B) COURT JUDGEMENTS — Lahore High Court (LHC), Sindh High Court (SHC), Supreme Court (SC), etc.

══════════════════════════════════════════════════════
CORE RULES — READ CAREFULLY BEFORE EVERY RESPONSE
══════════════════════════════════════════════════════

RULE 1 — ALWAYS USE RETRIEVED TEXT
You MUST base your answer ONLY on the retrieved passages provided.
Even if the question is general (e.g., "what to do if police asks for bribe"),
you MUST cite the specific law sections and articles from the retrieved passages.
DO NOT give generic advice without legal backing.

RULE 2 — INLINE CITATIONS ARE MANDATORY
Every single piece of advice or fact MUST end with an inline citation:
  • Law  →  [PPC, Section 161, Page 45]  or  [Constitution, Article 9, Page 12]
  • Case →  [LHC 2026, Khan vs State, Page 3]
If you cannot find a retrieved passage to support a statement, do not make it.

RULE 3 — PRACTICAL + LEGAL TOGETHER
For scenario questions ("what should I do if..."), give step-by-step practical
advice AND cite the law that protects each step. Citizens need to know WHAT to
do AND WHY the law allows it.

RULE 4 — OUTPUT STRUCTURE IS MANDATORY
You must ALWAYS produce the output in EXACTLY this structure, in this order:

  ──────────────────────────────────────────────────
  📖 ANSWER
  ──────────────────────────────────────────────────
  [Your main answer with inline citations.
   For practical questions: numbered steps, each with a law citation.
   For direct questions: explain the law clearly with citations.]

  ──────────────────────────────────────────────────
  📌 KEY PASSAGES YOU CAN QUOTE
  ──────────────────────────────────────────────────
  List 2–4 short direct quotes from the retrieved text that the citizen can
  show to police or use in court. Format each as:
  > "[exact quote from retrieved text]"
  — [Source, Section/Article, Page]

  ──────────────────────────────────────────────────
  📚 STATUTORY LAW REFERENCES
  ──────────────────────────────────────────────────
  List EVERY law chunk you used. Format:
  • [Document Name] | Section/Article X | Page Y | Brief description

  ──────────────────────────────────────────────────
  ⚖️ COURT JUDGEMENT REFERENCES
  ──────────────────────────────────────────────────
  List EVERY judgement chunk you used. Format:
  • [Court Name] ([Code]) | [Year] | [Case Title] | File: [filename] | Page [Y]
  If NO judgements were retrieved, write: "No court judgements retrieved for this query."

  ──────────────────────────────────────────────────
  🔎 CONCLUSION
  ──────────────────────────────────────────────────
  One short paragraph: what the law says + what courts have decided +
  the most important practical takeaway for the citizen.

RULE 5 — HONESTY
Never invent section numbers, case names, or page numbers.
If a retrieved passage does not clearly support something, say:
"Verify this point with a qualified advocate."

RULE 6 — LANGUAGE
Write in clear, simple language. The reader may not be a lawyer.
Avoid excessive legal jargon. Explain terms when you use them.
══════════════════════════════════════════════════════"""


# ═══════════════════════════════════════════════════════
#  CONTEXT BUILDER
# ═══════════════════════════════════════════════════════

def _build_context(relevant: List[Dict]) -> str:
    """
    Build the context block fed to the LLM.
    Law chunks first (Constitution → PPC → other), judgements second.
    Each chunk is clearly labelled so the LLM can cite it precisely.
    """
    law_chunks  = [c for c in relevant if c.get("source_type") != "judgement"]
    judg_chunks = [c for c in relevant if c.get("source_type") == "judgement"]

    blocks = []

    # ── Statutory law ──────────────────────────────────
    if law_chunks:
        by_src: Dict[str, List] = {}
        for c in law_chunks:
            by_src.setdefault(c["source"], []).append(c)

        blocks.append("══ STATUTORY LAW (cite these as [Document, Section, Page]) ══")
        for src, clist in by_src.items():
            blocks.append(f"\n  ─── {src.upper()} ───")
            for c in clist:
                label = c.get("article") or "General Text"
                blocks.append(
                    f"  [CITE AS: {src}, {label}, Page {c['page']}]\n"
                    f"  {c['text']}"
                )

    # ── Court judgements ───────────────────────────────
    if judg_chunks:
        blocks.append("\n══ COURT JUDGEMENTS (cite these as [Court Code Year, Case, Page]) ══")
        for c in judg_chunks:
            court_name = c.get("court_name", c.get("court_code", "Court"))
            court_code = c.get("court_code", "")
            year       = c.get("year", "")
            case       = c.get("case_title", "Unknown Case")
            fname      = c.get("pdf_filename", c.get("source", ""))
            label      = c.get("article") or "General Text"
            cite_as    = f"{court_code} {year}, {case}, Page {c['page']}"
            blocks.append(
                f"\n  ─── {court_name} | {year} | {case} ───"
                f"\n  [CITE AS: {cite_as}]"
                f"\n  File: {fname} | {label} | Page {c['page']}"
                f"\n  {c['text']}"
            )

    if not blocks:
        return (
            "=== NO PASSAGES RETRIEVED ===\n"
            "No relevant law sections or judgements were found.\n"
            "You must tell the user you could not find specific legal text "
            "and recommend they consult a qualified advocate.\n"
            "==============================\n"
        )

    return (
        "=== RETRIEVED PAKISTANI LEGAL SOURCES ===\n"
        "Use ONLY these passages. Cite each one you use.\n"
        + "\n---\n".join(blocks)
        + "\n==========================================\n"
    )


# ═══════════════════════════════════════════════════════
#  PROMPT BUILDER
# ═══════════════════════════════════════════════════════

def _build_prompt(question: str, relevant: List[Dict]) -> str:
    context = _build_context(relevant)
    law_count  = sum(1 for c in relevant if c.get("source_type") != "judgement")
    judg_count = sum(1 for c in relevant if c.get("source_type") == "judgement")

    return f"""{_SYSTEM_PROMPT}

{context}

═══════════════════════════════════════════════════════
USER QUESTION: {question}
═══════════════════════════════════════════════════════

You have {law_count} statutory law passage(s) and {judg_count} court judgement passage(s) above.

IMPORTANT REMINDERS BEFORE YOU WRITE:
1. Read ALL retrieved passages above before writing a single word.
2. For EVERY piece of advice → cite the law passage that supports it.
3. For EVERY court precedent → cite the judgement passage.
4. Follow the EXACT output structure: 📖 ANSWER → 📌 KEY PASSAGES → 📚 LAW REFERENCES → ⚖️ JUDGEMENT REFERENCES → 🔎 CONCLUSION
5. The LAW REFERENCES and JUDGEMENT REFERENCES sections MUST list every source you cited inline.

Answer:"""


# ═══════════════════════════════════════════════════════
#  VERIFICATION PROMPT
# ═══════════════════════════════════════════════════════

_VERIFY_PROMPT = """You are a senior Pakistani legal expert and QA reviewer.

Review the following AI-generated legal answer for a citizen.

CHECK FOR:
1. Every factual/legal claim has an inline citation → if missing, mark NEEDS_CORRECTION
2. The 📚 STATUTORY LAW REFERENCES section is present and complete
3. The ⚖️ COURT JUDGEMENT REFERENCES section is present (even if it says "No judgements retrieved")
4. The 🔎 CONCLUSION is present
5. No invented section numbers, case names, or page numbers
6. If the answer says "I don't know" without citing retrieved text → mark NEEDS_CORRECTION

Rate the answer: EXCELLENT / GOOD / NEEDS_CORRECTION

If EXCELLENT or GOOD → output the answer UNCHANGED.
If NEEDS_CORRECTION → output the corrected version maintaining the same structure.

Original Question: {question}

Generated Answer:
{answer}

Output the final verified answer ONLY.
Last line must be exactly:
[QA_STATUS: EXCELLENT|GOOD|NEEDS_CORRECTION — verified by {verifier}]
"""


# ═══════════════════════════════════════════════════════
#  MAIN ENTRY POINT
# ═══════════════════════════════════════════════════════

def ask(user_question: str, show_sources: bool = True) -> str:
    print(f"\n⏳ Processing …")

    # Step 1 — expand query (scenario → legal concepts)
    legal_query = expand_query(user_question)

    # Step 2 — retrieve
    retrieved = retrieve_chunks(legal_query, original_question=user_question)

    # Step 3 — cross-check
    relevant = crosscheck_chunks(user_question, retrieved)

    law_count  = sum(1 for c in relevant if c.get("source_type") != "judgement")
    judg_count = sum(1 for c in relevant if c.get("source_type") == "judgement")
    print(f"📊 Using {law_count} law chunk(s) + {judg_count} judgement chunk(s)")

    # Step 4 — build prompt + generate draft
    prompt = _build_prompt(user_question, relevant)
    print("📝 Generating draft …")

    draft, gemini_ok = call_gemini(prompt, temperature=0.2, max_tokens=3000)
    if not gemini_ok:
        print("🔄 Falling back to Groq …")
        draft, groq_ok = call_groq(prompt, temperature=0.2, max_tokens=3000)
        if not groq_ok:
            raise RuntimeError("Both Gemini and Groq failed.")
        primary_used_gemini = False
    else:
        primary_used_gemini = True

    # Step 5 — cross-model verification
    final = _verify_answer(user_question, draft, primary_used_gemini)

    # ── Print to console ────────────────────────────────
    print("\n" + "=" * 70)
    print(f"❓ QUESTION:\n{user_question}")
    print("=" * 70)
    print(f"\n{final}")

    if show_sources and relevant:
        print("\n" + "-" * 70)
        print("📌 SOURCE CHUNKS USED:")
        for i, c in enumerate(relevant, 1):
            st  = c.get("source_type", "law")
            art = c.get("article") or "General Text"
            if st == "judgement":
                court = c.get("court_code","")
                year  = c.get("year","")
                case  = c.get("case_title","")[:50]
                label = f"[JUDGEMENT] {court} {year} | {case}"
            else:
                label = f"[LAW] {c['source']}"
            print(f"  {i}. {label} | {art} | p.{c['page']} | score={c['score']:.3f}")

    print("=" * 70 + "\n")
    return final


def _verify_answer(question: str, draft: str, primary_used_gemini: bool) -> str:
    verifier = "Groq" if primary_used_gemini else "Gemini"
    prompt   = _VERIFY_PROMPT.format(
        question=question, answer=draft, verifier=verifier
    )
    print(f"\n🔎 Verifying with {verifier} …")

    if primary_used_gemini:
        verified, ok = call_groq(prompt, temperature=0.1, max_tokens=3500)
        if ok:
            print("✅ Verification complete (Groq).")
            return verified
        return draft

    verified, ok = call_gemini(prompt, temperature=0.1, max_tokens=3500)
    if ok:
        print("✅ Verification complete (Gemini).")
        return verified

    # Both down — Groq self-review
    print("⚠️  Gemini unavailable — Groq self-review …")
    sr_prompt = prompt.replace(f"verified by {verifier}", "self-reviewed by Groq")
    verified, ok = call_groq(sr_prompt, temperature=0.1, max_tokens=3500)
    return verified if ok else draft


print("✅ Answer generator ready.")
````

````python
# ════════════════════════════════════════════════════════
#  MAIN  —  Flask API server (Google Colab + Cloudflare Tunnel)
#  No account or token required.
#  Exposes POST /ask  →  { "question": "..." }
#                    ←  { "answer": "...", "sources": [...] }
# ════════════════════════════════════════════════════════

import os, threading, subprocess, time, re, pdfplumber
from pathlib import Path
from flask import Flask, request, jsonify
from flask_cors import CORS


# ── Debug helpers ─────────────────────────────────────────

def debug_section(section_number: str):
    found = [c for c in _chunks if c.get("article") and section_number in c["article"]]
    if found:
        print(f"\n✅ {len(found)} chunk(s) contain Section/Article '{section_number}':")
        for c in found:
            st = "(judgement)" if c.get("source_type") == "judgement" else "(law)"
            print(f"   {st} [{c['source']}] {c['article']} | p.{c['page']}")
            print(f"   Preview: {c['text'][:100].replace(chr(10),' ')} …\n")
    else:
        print(f"\n⚠️  No chunks tagged with '{section_number}'.")


def debug_court(court_code: str):
    found = [c for c in _chunks if c.get("court_code","").upper() == court_code.upper()]
    years = sorted({c.get("year","") for c in found})
    cases = sorted({c.get("case_title","") for c in found})[:10]
    print(f"\n📋 {court_code.upper()} — {len(found)} chunks | Years: {years}")
    print(f"   Sample cases: {cases}")


def debug_stats():
    law_chunks  = [c for c in _chunks if c.get("source_type") != "judgement"]
    judg_chunks = [c for c in _chunks if c.get("source_type") == "judgement"]
    courts = {}
    for c in judg_chunks:
        key = f"{c.get('court_code','')} {c.get('year','')}"
        courts[key] = courts.get(key, 0) + 1
    print(f"\n📊 INDEX STATS")
    print(f"   Law chunks       : {len(law_chunks)}")
    print(f"   Judgement chunks : {len(judg_chunks)}")
    print(f"   Breakdown by court/year:")
    for k, v in sorted(courts.items()):
        print(f"      {k} — {v} chunks")


# ── Startup ───────────────────────────────────────────────

index, embedder, _chunks = build_or_update_index()
set_retrieval_state(index, embedder, _chunks)
debug_stats()
print("\n🔍 Sanity-check Section 295 …")
debug_section("295")


# ── Flask app ─────────────────────────────────────────────

app = Flask(__name__)
CORS(app)


@app.route("/", methods=["GET"])
def health():
    law_count  = sum(1 for c in _chunks if c.get("source_type") != "judgement")
    judg_count = sum(1 for c in _chunks if c.get("source_type") == "judgement")
    return jsonify({
        "status"          : "ok",
        "law_chunks"      : law_count,
        "judgement_chunks": judg_count,
    })


@app.route("/ask", methods=["POST"])
def ask_endpoint():
    """
    POST /ask
    Body (JSON): { "question": "...", "show_sources": true }
    Response   : { "answer": "...", "sources": [...] }
    """
    data = request.get_json(silent=True)
    if not data or not data.get("question", "").strip():
        return jsonify({"error": "Missing 'question' field."}), 400

    user_question = data["question"].strip()
    show_sources  = data.get("show_sources", True)

    try:
        legal_query = expand_query(user_question)
        retrieved   = retrieve_chunks(legal_query, original_question=user_question)
        relevant    = crosscheck_chunks(user_question, retrieved)

        law_count  = sum(1 for c in relevant if c.get("source_type") != "judgement")
        judg_count = sum(1 for c in relevant if c.get("source_type") == "judgement")
        print(f"📊 Using {law_count} law chunk(s) + {judg_count} judgement chunk(s)")

        answer = ask(user_question, show_sources=False)

        sources = []
        if show_sources:
            for c in relevant:
                st  = c.get("source_type", "law")
                art = c.get("article") or "General Text"
                if st == "judgement":
                    sources.append({
                        "type"      : "judgement",
                        "court"     : c.get("court_name", c.get("court_code", "")),
                        "court_code": c.get("court_code", ""),
                        "year"      : c.get("year", ""),
                        "case"      : c.get("case_title", ""),
                        "file"      : c.get("pdf_filename", c.get("source", "")),
                        "article"   : art,
                        "page"      : c["page"],
                        "score"     : round(c["score"], 3),
                    })
                else:
                    sources.append({
                        "type"   : "law",
                        "source" : c["source"],
                        "article": art,
                        "page"   : c["page"],
                        "score"  : round(c["score"], 3),
                    })

        return jsonify({"answer": answer, "sources": sources})

    except Exception as e:
        print(f"❌ Error: {e}")
        return jsonify({"error": str(e)}), 500


@app.route("/stats", methods=["GET"])
def stats_endpoint():
    law_chunks  = [c for c in _chunks if c.get("source_type") != "judgement"]
    judg_chunks = [c for c in _chunks if c.get("source_type") == "judgement"]
    courts = {}
    for c in judg_chunks:
        key = f"{c.get('court_code','')} {c.get('year','')}"
        courts[key] = courts.get(key, 0) + 1
    return jsonify({
        "law_chunks"      : len(law_chunks),
        "judgement_chunks": len(judg_chunks),
        "courts"          : courts,
    })


# ── Cloudflare Tunnel ─────────────────────────────────────

def install_cloudflared():
    """Download cloudflared binary if not already present."""
    if not Path("/usr/local/bin/cloudflared").exists():
        print("⬇️  Installing cloudflared …")
        os.system(
            "wget -q https://github.com/cloudflare/cloudflared/releases/latest/"
            "download/cloudflared-linux-amd64 -O /usr/local/bin/cloudflared && "
            "chmod +x /usr/local/bin/cloudflared"
        )
        print("✅ cloudflared installed.")
    else:
        print("✅ cloudflared already present.")


def start_cloudflare_tunnel(port: int = 5000) -> tuple:
    """
    Start a Cloudflare Quick Tunnel (no account needed).
    Parses the trycloudflare.com URL from stderr and returns (url, process).
    """
    proc = subprocess.Popen(
        ["cloudflared", "tunnel", "--url", f"http://localhost:{port}"],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
    )

    url_pattern = re.compile(r"https://[\w\-]+\.trycloudflare\.com")
    public_url  = None
    deadline    = time.time() + 30   # wait up to 30 s

    while time.time() < deadline:
        line = proc.stderr.readline().decode("utf-8", errors="ignore")
        if not line:
            time.sleep(0.2)
            continue
        m = url_pattern.search(line)
        if m:
            public_url = m.group(0)
            break

    if not public_url:
        raise RuntimeError("Cloudflare tunnel did not start in time. Check output above.")

    return public_url, proc


# ── Start everything ──────────────────────────────────────

print("\n" + "=" * 70)
print("🇵🇰  Pakistan Legal Assistant  —  API Server Mode")
print("=" * 70)

# 1. Install cloudflared (no account / token needed)
install_cloudflared()

# 2. Start Flask in a background thread
flask_thread = threading.Thread(
    target=lambda: app.run(host="0.0.0.0", port=5000, use_reloader=False),
    daemon=True
)
flask_thread.start()
time.sleep(1)   # let Flask bind

# 3. Open Cloudflare tunnel
print("🚇 Starting Cloudflare tunnel …")
public_url, cf_proc = start_cloudflare_tunnel(5000)

print(f"\n{'=' * 70}")
print(f"🌐 PUBLIC API URL: {public_url}")
print(f"\n   Endpoints:")
print(f"   GET  {public_url}/         → health check")
print(f"   POST {public_url}/ask      → ask a legal question")
print(f"   GET  {public_url}/stats    → index statistics")
print(f"\n   Example curl:")
print(f'   curl -X POST {public_url}/ask \\')
print(f'        -H "Content-Type: application/json" \\')
print(f"        -d '{{\"question\": \"What are my rights if arrested?\"}}'")
print(f"\n⚠️  Keep this cell running — closing it kills the server.")
print("=" * 70)

# 4. Keep the main thread alive
try:
    while True:
        time.sleep(60)
except KeyboardInterrupt:
    cf_proc.terminate()
    print("\n🛑 Server stopped.")
````
